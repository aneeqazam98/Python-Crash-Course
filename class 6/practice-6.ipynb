## Dictionaries

```python
# list 
marks = [100,90,80,76,57]
subjects = ["maths","Sci","python"]
print(marks[2])
```

```python
# dictionaries

# key - value pairs
# student - ayan 

# name  - "ayan"
# age   - 50
# marks - 100
# email - ayanhussain@gmail.com
# phone_no - 0000000000
```

```python
# syntax -- starts with curly braces {}

student = {
    "name":"ayan",
    "age": 50,
    "marks": 100,
    "email": "ayanhussain@gmail.com"
}
print(student)
```

```python

print(type(student))
```

```python

print(student["email"])
```

```python


print(student["roll_no"])
```

```python
student = {
    "name":"ayan",
    "age": 50,
    "marks": 100,
    "email": "ayanhussain@gmail.com",
    "passed":True,
    "grades":[89,67,90,23,45]
}
student["passed"] = False
print(student)
```

```python
# dict["key"] = value
# if key exists then update the value then it will add an new key 
student["id"] = 'ABC-123'

print(student)
```
