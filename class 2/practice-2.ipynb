# Operators


```python
A = int(input("enter some number: "))
B = int(input("enter another number: "))
print(A + B) # addition
print(A - B) # subtraction
print(A * B) # multiplication
print(A / B) # division
print(A // B) # floor division
print(A % B) # modulus
print(A ** B) # exponentiation
```

```python
# STADIUM TAKE AGE INPUT AND DECIDE THE PRICE OF THE TICKET 
age = int(input("enter your age: "))
print(age)
if age <= 5 or age > 60:
    print("you can enter for free")
elif age >= 5 and age <= 18:
    print("you need to pay 10$")
```

```python
# convert age in days,hours,minutes,seconds
age = int(input("enter your age: "))
days = age * 365
hours = days * 24
minutes = hours * 60
seconds = minutes * 60
print("you have been alive for", days, "days")
print("you have been alive for", hours, "hours")
print("you have been alive for", minutes, "minutes")
print("you have been alive for", seconds, "seconds")
```
