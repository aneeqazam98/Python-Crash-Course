```python
def introduce(name, age, city):
    print(f"{name} is , age is {age},city is {city }")

# positional argurments -- order matters
introduce("ayan",50,"karachi")

# keyword arguments -- order does not matter
introduce(name="Ayan",city="ISB",age=12)
```

```python
# Write get_grade(score) that returns a letter grade.
# Then loop over a list of students and print each result.

def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

students = [
    {"name": "Ali", "score": 92},
    {"name": "Sara", "score": 75},
    {"name": "Ahmed", "score": 55},
    {"name": "Zara", "score": 88}
]

for student in students:
    grade = get_grade(student["score"])
    print(f"{student['name']} got grade {grade}")
```

```python
# instead of creating multiple functions, we can do this in one and can return multiple values 

# multiple returns 
def get_stats(numbers):
    return max(numbers) , min(numbers), sum(numbers)


numbers = [5,2,4,7,9,10,12,2,3] # min max sum
max_no , min_n0, sum_no = get_stats(numbers)
print(max_no , min_n0, sum_no)

```
