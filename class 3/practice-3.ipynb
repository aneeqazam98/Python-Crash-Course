## If- Else


```python
# Based on conditions 
# kids youtube age limit 
# greater then 10 and less 18 

# Syntax 

# if condition:
#     code logic
# else:
#     code logic 
```

```python
age = int(input("Enter your age"))

if age > 10 and age < 18 :
    print("Yes! your are eligible....")
else:
    print("No! you are not eligible...")
```

```python
# Students Grade Calculations 

marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade A+")
elif marks >= 80  and marks < 90:
    print("Grade: A")
elif marks >= 70 and marks < 80:
    print("Grade B")
else:
    print("Grade F")
```

```python
# checking a number 
# if positive number then print positive or print negative else print zero 

number = int(input("enter a number"))

if number > 0:
   print("positive number")

elif number < 0:
   print("negative number")

else:
    print("Zero")
```

```python
# Nested If - Else

# we have logging systesm
# 1 for students 
# 2 for teachers

logged_in = False
is_teacher = False

if logged_in:
    print("Logged In")

    if is_teacher:  # nested if 
        print("welcome, teacher!")
    else: 
        print("Welcome, Student")

else: # first if ka else hai
    print("Please Login ")
```
