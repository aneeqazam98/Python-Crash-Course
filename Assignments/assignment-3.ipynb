### List & String Logic

```python
# 1. 
nums = [10, 20, 30, 40, 50, 60]
print("1. Alternate elements:")
for i in range(0, len(nums), 2):

    print(nums[i])
```

```python
# 2. 
items = [1, 2, 3, 4, 5]
reversed_list = []
for i in range(len(items) - 1, -1, -1):
    reversed_list.append(items[i])
print("\n2. Reversed list:", reversed_list)
```

```python
# 3.
numbers = [23, 89, 12, 56, 78]
largest = numbers[0]
for n in numbers:
    if n > largest:
        largest = n
print("\n3. Largest number is:", largest)
```

```python
# 4. 

list_to_rotate = [1, 2, 3, 4, 5]
last_element = list_to_rotate.pop() # Removes 5
list_to_rotate.insert(0, last_element) # Puts 5 at index 0
print("\n4. Rotated list:", list_to_rotate)
```

```python
# 5. 
text = "I love apple and apple is healthy"
word_to_delete = "apple"
words = text.split()
filtered_words = []
for w in words:
    if w != word_to_delete:
        filtered_words.append(w)
result_string = " ".join(filtered_words)
print("\n5. String after deletion:", result_string)
```

```python
# 6.
date_input = "03/12/2021"
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]

m, d, y = date_input.split('/')
month_name = months[int(m) - 1]
print(f"\n6. Formatted Date: {month_name} {d}, {y}")
```

```python
# 7.
sentence = "stop and smell the roses"
words = sentence.split()
capitalized_words = []
for w in words:
    capitalized_words.append(w.capitalize())
print("\n7. Capitalized:", " ".join(capitalized_words))
```

```python
# 8.
matrix = [
    [2, 11, 7, 12],
    [5, 2, 9, 15],
    [8, 3, 10, 42]
]
print("\n8. Matrix Row Sums:")
row_count = 1
for row in matrix:
    print(f"Sum of row {row_count} = {sum(row)}")
    row_count += 1
```

```python
# 9. 
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
result_add = [[0, 0], [0, 0]]

for i in range(len(A)):
    for j in range(len(A[0])):
        result_add[i][j] = A[i][j] + B[i][j]
print("\n9. Sum of two matrices:", result_add)
```

```python
## 10.
X = [[1, 2], [3, 4]]
Y = [[5, 6], [7, 8]]
result_mult = [[0, 0], [0, 0]]

for i in range(len(X)):
    for j in range(len(Y[0])):
        for k in range(len(Y)):
            result_mult[i][j] += X[i][k] * Y[k][j]
print("\n10. Product of two matrices:", result_mult)
```
