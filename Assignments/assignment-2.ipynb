```python
## bonus calculator

salary = float(input("Enter your salary: "))
years = int(input("Enter years of service: "))
if years > 5:
    bonus = salary * 0.05
else:
    bonus = 0

print(f"Net Bonus Amount: {bonus}")
```

```python
## discount calculator
quantity = int(input("Enter quantity: "))
total_cost = quantity * 100
if total_cost > 1000:
    discount = total_cost * 0.10
    total_cost -= discount
print("Total cost:", total_cost)
```

```python
## grade calculator
marks = int(input("Enter marks: "))
if marks < 25:
    print("Grade F")
elif marks < 45:
    print("Grade E")
elif marks < 50:
    print("Grade D")
elif marks < 60:
    print("Grade C")
elif marks < 80:
    print("Grade B")
else:
    print("Grade A")
```

```python
## attendance percentage calculator
held = int(input("Enter classes held: "))
attended = int(input("Enter classes attended: "))
percentage = (attended / held) * 100
print(f"Attendance: {percentage:.2f}%")

if percentage >= 75:
    print("Allowed in exam")
else:
    print("Not allowed in exam")
```

```python
## attendance with medical condition
held = int(input("Enter classes held: "))
attended = int(input("Enter classes attended: "))
medical = input("Medical cause? (Y/N): ")
percentage = (attended / held) * 100

print(f"Attendance: {percentage:.2f}%")

if percentage >= 75 or medical == 'Y':
    print("Allowed in exam")
else:
    print("Not allowed in exam")
```

```python
## leap year checker
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
## electricity bill calculator
units = int(input("Enter units consumed: "))

if units <= 100:
    bill = 0
elif units <= 300:
    bill = (units - 100) * 5
else:
    bill = (200 * 5) + (units - 300) * 10
print("Total bill:", bill)
```

```python
## oldest and youngest 
a = int(input("Enter age of person 1: "))
b = int(input("Enter age of person 2: "))
c = int(input("Enter age of person 3: "))

oldest = max(a, b, c)
youngest = min(a, b, c)

print("Oldest:", oldest)
print("Youngest:", youngest)
```
